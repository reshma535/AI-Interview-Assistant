```python
from transformers import pipeline
```

```python
generator = pipeline("text-generation", model="distilgpt2")
```

```python
result = generator("Hello, I am your AI interviewer", max_length=30)

print(result)
```

```python
role = "Java Developer"

answer = """
I know Java, OOP concepts, collections, and built Java projects.
"""

prompt = f"""
You are an interviewer hiring for {role}.

Candidate answer:
{answer}

Give interview feedback:
"""
```

```python
result = generator(prompt, max_length=100)

print(result[0]['generated_text'])
```
